```python
# Importar librerías de visualización
import matplotlib.pyplot as plt
import seaborn as sns
 
# Configuración de estilo
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 4)
```

```python
for i_grafica in ['FechaInscripcionBeneficiario','CantidadDeBeneficiarios']:
    sns.histplot(data=df_FAMILIAS_ACCION, x=i_grafica, kde=True)
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_FAMILIAS_ACCION, x='TipoPoblacion', hue='Genero', palette='Set2')
 
plt.title('TipoPoblacion vs Genero')
plt.xlabel('TipoPoblacion')
plt.ylabel('CantidadDeBeneficiarios')
plt.xticks(rotation=45)
plt.legend(title='Genero')
plt.tight_layout()
plt.show()
 
```
